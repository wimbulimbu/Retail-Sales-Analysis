```python
import pandas as pd
import sqlite3
import plotly.express as px
```

```python
conn = sqlite3.connect('sales.db')
```

```python
# Loading retail sales data set from Kaggle

retail_s = pd.read_csv('data/retail_sales_dataset.csv')
```

```python
'''
Findings:
- 1000 tansactions.
- No duplicated rows found.
- No missing values found.
- Date column is in string format so that needs to be changed to date time.
- Column names have spaces.

'''
```

```python
retail_s['Date'] = pd.to_datetime(retail_s['Date'])
retail_s.columns = retail_s.columns.str.replace(' ', '_')
```

```python
retail_s.dtypes
```

```python
retail_s.head()
```

```python
retail_s['month'] = retail_s['Date'].dt.month_name()
retail_s['day'] = retail_s['Date'].dt.day_name()
retail_s['month_val'] = retail_s['Date'].dt.month
retail_s['day_val'] = retail_s['Date'].dt.day_of_week
```

```python
bins = [0, 25, 35, 45, 55, 100]

labels = [
    '18-25',
    '26-35',
    '36-45',
    '46-55',
    '56+'
]

retail_s['Age_group'] = pd.cut(retail_s['Age'],
                               bins=bins,
                               labels=labels)
```

```python
retail_s
```

```python
'''
Changes made: 

- Date column has been changed to date time object.
- Column name with spaces have been replaced with '_'.
- Created new columns for day and month name including its numerical value for easier analysis.
- Created new columns for age group for easoer analysis later.

'''
```

```python
retail_s.to_sql('initial_data', conn, if_exists='replace', index=False)
```

```python
retail_sales = pd.read_sql('''
SELECT * 
FROM initial_data
''', conn,
parse_dates=['Date'])

retail_sales.head()
```

```python
# This section explores sales performance across revenue, product categories, customer demographics, and time-based trends.
```

```python
# Revenue insights
total_revenue = retail_sales['Total_Amount'].sum()
average_order_val = retail_sales['Total_Amount'].mean()
largest_sale = retail_sales['Total_Amount'].max()
```

```python
'''
Findings:

- Total revenue generated : 456000
- Average order value : 456
- Largest sale : 2000

'''
```

```python
# product insights
product_categories = retail_sales.groupby(['Product_Category'], as_index=False).agg(
    Total_revenue = ('Total_Amount', 'sum'),
    Total_quantity = ('Quantity', 'sum')
).sort_values(by='Total_revenue', ascending=False)
```

```python
product_categories
```

```python
'''
Findings:

- Electronics generated the most revenue.
- Clothing sold the most quantity.

'''
```

```python
# Time insights
monthly = retail_sales.groupby(['month_val'], as_index=False).agg(
    revenue = ('Total_Amount', 'sum')
).sort_values(by='revenue', ascending=False).reset_index(drop=True)

daily = retail_sales.groupby(['day_val'], as_index=False).agg(
    revenue = ('Total_Amount', 'sum')
).sort_values(by='revenue', ascending=False).reset_index(drop=True)
```

```python
#sql queries

monthly_revenue = pd.read_sql('''
SELECT month_val, month, SUM(Total_Amount) as revenue
FROM initial_data
GROUP BY month_val, month
ORDER BY month_val ASC
                                     ''', conn)

category_revenue = pd.read_sql('''
SELECT Product_Category, SUM(Total_Amount) as revenue
FROM initial_data
GROUP BY Product_Category
ORDER BY revenue DESC                            
                             ''', conn)

Top_customers = pd.read_sql('''
SELECT Customer_ID, Total_Amount
FROM initial_data
ORDER BY Total_Amount DESC
LIMIT 10
                            ''', conn)

# Due to each customer appearing once, analysis was limited

gender_revenue = pd.read_sql('''
SELECT Gender, SUM(Total_Amount) as Revenue
FROM initial_data
GROUP BY Gender
ORDER BY Revenue DESC
                            ''', conn)

age_group_revenue = pd.read_sql('''
SELECT Age_group, SUM(Total_Amount) as Revenue
FROM initial_data
GROUP BY Age_group
ORDER BY Revenue DESC
                            ''', conn)

average_order_by_category = pd.read_sql('''
SELECT Product_Category, AVG(Quantity) as average_order
FROM initial_data
GROUP BY Product_Category
ORDER BY average_order DESC                                       
                                        ''', conn)
```
